```python
import re

def extract_https(input_string):
    pattern = '(?P<group1>((https:\/\/openalex.org\/W[0-9]{10})))'

    matches = re.finditer(pattern, input_string)# ???
    
    result1=[match.group('group1') for match in matches ]
    return(set(result1))
```

```python
ddc_daten={str(sublist_key)+'.'+str(i): element for sublist_key, sublist in list_id.items() 
          for element,i in zip(extract_https(sublist),range(len(extract_https(sublist))))}
```
